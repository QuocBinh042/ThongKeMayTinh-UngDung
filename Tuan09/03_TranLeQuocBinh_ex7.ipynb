## Kiểm định đặc trưng tỉ lệ (p)

### Bài 1

Dataset 18 (__M&M__) chứa thông tin trọng lượng và màu sắc của mẫu các viên kẹo của hãng M&M. 

Sử dụng mẫu trên bạn hãy kiểm tra phát biểu sau: __"20% kẹo M&M có màu đỏ"__ với mức ý nghĩa là 5%

*Hướng dẫn: tính tỷ lệ các viên kẹo có màu đỏ từ dữ liệu, sử dụng kết quả để kiểm định phát biểu trên*

```python
# import library
import numpy as np
from scipy.stats import norm
import math 
import pandas as pd
```

```python
df = pd.read_excel('./data/18_M_M.xls')
df 
```

```python
# Tính tỷ lệ màu đỏ từ dữ liệu
red_ratios = df['Red'].dropna()
red_ratio_mean = red_ratios.mean()

# Giả thuyết Ho: 20% kẹo M&M có màu đỏ
p_0 = 0.20

# Tính toán giá trị kiểm định z-score
z_score = (red_ratio_mean - p_0) / (np.std(red_ratios, ddof=1) / np.sqrt(len(red_ratios)))

# Tính toán p-value
p_value = 1 - norm.cdf(z_score)  

# In ra kết quả
print("Tỷ lệ các viên kẹo có màu đỏ từ dữ liệu mẫu:", red_ratio_mean)
print("Giá trị z-score:", z_score.round(4))
print("p-value:", p_value.round(4))
```

```python
# Kiểm tra xem có bác bỏ giả thuyết không
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: Tỷ lệ kẹo M&M có màu đỏ không phải là 20%.")
else:
    print("Fail to reject null hypothesis: Tỷ lệ kẹo M&M có màu đỏ có thể là 20%."
```

### Bài 2

Dataset 3 (__FRESH15__) chứa thông tin khảo sát những thay đổi về trọng lượng (weight - WT) và chỉ số khối của cơ thể (BMI) của sinh viên năm nhất tại một trường Đại Học. Thời điểm khảo sát là bắt đầu học kỳ (tháng 9 - SEP) và kết thúc học kỳ (tháng 4 - APR), các sinh viên được chọn ngẫu nhiên trong trường.

Sử dụng mẫu trên, hãy kiểm tra phát biểu:__"50% số sinh viên trong trường là nam"__ với mức ý nghĩa 50%

Các thuộc tính trong bộ dữ liệu:
- SEX: Giới tính
- WTSEP: trọng lượng đo vào tháng 9 (mới nhập học)
- WTAPR: trọng lượng đo vào tháng 4 (kết thúc năm học)
- BMISP: BMI tháng 9
- BMIAP: BMI tháng 4

```python
df = pd.read_excel('./data/03_FRESH15.xls')
df 
```

```python
# Tính tỷ lệ nam sinh viên từ dữ liệu mẫu
male_count = df[df['SEX'] == 'M'].shape[0]
total_count = df.shape[0]
male_ratio = male_count / total_count

# Giả thuyết Ho: tỷ lệ nam sinh viên là 50%
p_0 = 0.50

# Tính toán giá trị z-score
z_score = (male_ratio - p_0) / np.sqrt(p_0 * (1 - p_0) / total_count)

# Tính toán p-value
p_value = 2 * (1 - norm.cdf(np.abs(z_score))) 

# In kết quả
print("Tỷ lệ nam sinh viên từ dữ liệu mẫu:", male_ratio)
print("Giá trị z-score:", z_score.round(4))
print("p-value:", p_value.round(4))
```

```python
# Kiểm tra xem có bác bỏ giả thuyết không
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: Không có 50% số sinh viên trong trường là nam.")
else:
    print("Fail to reject null hypothesis: Có thể có 50% số sinh viên trong trường là nam.")
```

### Bài 3

Dataset 6(__BEARS__) chứa thông tin khảo sát ngẫu nhiên các con gấu trong một khu rừng.

Sử dụng mẫu trên, bạn hãy kiểm tra phát biểu: __"tỷ lệ gấu đực trong khu rừng là 50%""__ với mức ý nghĩa 90%.

Biết rằng thuộc tính SEX có hai giá trị: 1 - gấu đực, 2- gấu cái.

```python
df = pd.read_excel('./data/06_BEARS.xls')
df 
```

```python
# Tính tỷ lệ gấu đực từ dữ liệu mẫu
male_count = df[df['SEX'] == 1].shape[0]
total_count = df.shape[0]
male_ratio = male_count / total_count

# Giả thuyết Ho: Tỷ lệ gấu đực trong khu rừng là 50%
p_0 = 0.50

# Tính toán giá trị kiểm định z-score
z_score = (male_ratio - p_0) / np.sqrt((p_0 * (1 - p_0)) / total_count)

# Tính toán p-value
p_value = 1 - norm.cdf(z_score)  # one-tailed test

# In ra kết quả
print("Tỷ lệ gấu đực từ dữ liệu mẫu:", male_ratio)
print("Giá trị z-score:", z_score.round(4))
print("p-value:", p_value.round(4))
```

```python
# Kiểm tra xem có bác bỏ giả thuyết không
alpha = 0.10  # Mức ý nghĩa 90%
if p_value < alpha:
    print("Reject null hypothesis: Tỷ lệ gấu đực trong khu rừng không phải là 50%.")
else:
    print("Fail to reject null hypothesis: Tỷ lệ gấu đực trong khu rừng có thể là 50%.")
```

### Bài 4

Dataset 9 __(MOVIES)__ chứa thông tin khảo sát một số phim trong khoảng thời gian 33 năm gần đây.

Có ý kiến cho rằng: __"55% các phim trong khoảng thời gian 33 năm gần đây có nhãn R (là phim giới hạn độ tuổi - Restricted)"__ .Theo bạn với mức ý nghĩa 95% ý kiến trên có đúng không?

Tham khảo: [Hệ thống phân loại của hiệp hội điện ảnh Mỹ](https://vi.wikipedia.org/wiki/H%E1%BB%87_th%E1%BB%91ng_ph%C3%A2n_lo%E1%BA%A1i_phim_c%E1%BB%A7a_MPAA#Ph%C3%A2n_lo%E1%BA%A1i_v%E1%BB%9Bi_c%C3%A1c_trailer_phim)

```python
df = pd.read_excel('./data/09_MOVIES.xls')
df 
```

```python
# Tính tỷ lệ phim có nhãn R từ dữ liệu mẫu
r_count = df[df['Rating'] == 'R'].shape[0]
total_count = df.shape[0]
r_ratio = r_count / total_count

# Giả thuyết Ho: 55% các phim trong khoảng thời gian 33 năm gần đây có nhãn R
p_0 = 0.55

# Tính toán giá trị kiểm định z-score
z_score = (r_ratio - p_0) / np.sqrt((p_0 * (1 - p_0)) / total_count)

# Tính toán p-value
p_value = 1 - norm.cdf(z_score)  # one-tailed test

# In ra kết quả
print("Tỷ lệ phim có nhãn R từ dữ liệu mẫu:", r_ratio)
print("Giá trị z-score:", z_score.round(4))
print("p-value:", p_value.round(4))
```

```python
# Kiểm tra xem có bác bỏ giả thuyết không
alpha = 0.05  # Mức ý nghĩa 95%
if p_value < alpha:
    print("Reject null hypothesis: Tỷ lệ phim có nhãn R không phải là 55%.")
else:
    print("Fail to reject null hypothesis: Tỷ lệ phim có nhãn R có thể là 55%.")
```

### Bài 5

Dataset 13 __(VOLTAGE)__ liệt kê các mức điện áp được đo tại các hộ gia đình bởi các máy đo khác nhau: tại nhà (HOME), tại máy phát độc lập (Generator), tại trạm phát (UPS).

Công ty cung cấp điện cho vùng này cam kết rằng: __"Mục tiêu cung cấp điện cho mỗi hộ gia đình là 120V"__ . Biết rằng độ lệch chuẩn của tất cả các mức điện áp đo được là 0.24V.

Với mức ý nghĩa 1%. Bạn hãy kiểm tra xem công ty có làm đúng cam kết hay không?

```python
df = pd.read_excel('./data/13_VOLTAGE.xls')
df 
```

```python
# Độ lệch chuẩn đã cho
sigma = 0.24

# Trung bình mẫu
mean_voltage_home = df['Home'].mean()

# Số lượng quan sát
n = len(df)

# Giả thuyết Ho: Mức điện áp trung bình là 120V
mu_0 = 120

# Tính toán giá trị z-score
z_score = (mean_voltage_home - mu_0) / (sigma / np.sqrt(n))

# Tính toán p-value
p_value = 1 - norm.cdf(z_score)


# In ra kết quả
print("Trung bình mức điện áp tại nhà: {:.2f}V".format(mean_voltage_home))
print("Giá trị z-score: {:.2f}".format(z_score))
print("p-value: {:.6f}".format(p_value))
```

```python
# Mức ý nghĩa
alpha = 0.01

# Kiểm tra xem có bác bỏ giả thuyết không
if p_value < alpha:
    print("Bác bỏ giả thuyết Ho: Công ty không làm đúng cam kết.")
else:
    print("Không đủ bằng chứng để bác bỏ giả thuyết Ho: Công ty có thể đã làm đúng cam kết.")
```

### Bài 7

Dataset 19(__SCREWS__) chứa thông tin khảo sát chiều dài một số đinh ốc được cung cấp bởi công ty Crown Bolt (đơn vị inches)

Trên bao bì của các đinh ốc do công ty sản xuất có ghi chiều dài của các đinh ốc là 3/4 inches. Sử dụng mẫu trên bạn hãy cho biết chiều dài của đinh ốc lấy từ mẫu thu nhận được từ dữ liệu trên có nhất quán với chiều dài ghi trên bao bì hay không? Kiểm định với mức ý nghĩa 5%

```python
df = pd.read_excel('./data/19_SCREWS.xls')
df 
```

```python
# Giả thuyết Ho: Chiều dài trung bình của đinh ốc là 3/4 inches
mu_0 = 3/4

# Tính trung bình mẫu
sample_mean = df.mean()

# Tính độ lệch chuẩn mẫu
sample_std = np.std(df, ddof=1)

# Số lượng quan sát
n = len(df)

# Tính giá trị t-score
t_score = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Tính p-value
p_value = norm.sf(np.abs(t_score)) * 2

# In kết quả
print("Trung bình mẫu của chiều dài đinh ốc:", round(sample_mean, 4), "inches")
print("Độ lệch chuẩn mẫu của chiều dài đinh ốc:", round(sample_std, 4), "inches")
print("Giá trị t-score:", round(t_score, 4))
print("p-value:", p_value)
```

```python
# Mức ý nghĩa
alpha = 0.05

# Kiểm tra xem có bác bỏ giả thuyết không
if p_value < alpha:
    print("Bác bỏ giả thuyết Ho: Chiều dài của đinh ốc không nhất quán với chiều dài ghi trên bao bì.")
else:
    print("Không đủ bằng chứng để bác bỏ giả thuyết Ho: Chiều dài của đinh ốc có thể nhất quán với chiều dài ghi trên bao bì.")
```
