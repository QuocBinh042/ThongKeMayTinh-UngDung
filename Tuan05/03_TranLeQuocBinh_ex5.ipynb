### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:
a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.
b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.
c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?
d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
%matplotlib inline
import numpy as np
import math
from scipy import stats as stats
import matplotlib.pyplot as plt
```

```python
POP=np.random.randint(120,201, size=10000)
print(POP)
mean_POP=np.mean(POP)
print('Chiều cao trung bình của Quần Thể : {}'.format(mean_POP))
std_POP=np.std(POP)
print("Độ lệch chuẩn của quần thể là:",std_POP)
```

```python
tyle_cao = np.count_nonzero(POP >= 180) / len(POP)
print("Tỷ lệ người cao trong quần thể là:", tyle_cao)
```

```python
plt.hist(POP, bins=20, rwidth=0.8)
plt.xticks(range(120, 201, 10))
plt.xlabel('Chiều cao')
plt.ylabel('Số lượng')
plt.title('Biểu đồ histogram về chiều cao của quần thể')
plt.show()

#trung vi của quần thể
median_POP=np.median(POP)
# Dự đoán về phân phối chuẩn của quần thể
if np.abs(mean_POP - median_POP) < 0.1 * std_POP:
    print("Quần thể có phân phối chuẩn.")
else:
    print("Quần thể không có phân phối chuẩn.")
```

```python
plt.hist(POP, bins=20, density=True, alpha=0.5, color='blue')

x = np.linspace(mean_POP - 3*std_POP, mean_POP + 3*std_POP, 100)
plt.plot(x, stats.norm.pdf(x, loc=mean_POP, scale=std_POP), color='red')
plt.xticks(range(120, 201, 10))
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.title('Histogram and Density Plot of Population Heights')

plt.show()
```

### Bài 5: Khảo sát tỷ lệ mẫu

a. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram tỷ lệ người cao của các mẫu
a. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 500. Vẽ histogram tỷ lệ người cao của các mẫu

```python
# Tạo mẫu chiều cao của dân số
heights = np.random.randint(120,201,size=1000)
# Tính tỷ lệ người cao trong mỗi mẫu
def get_tyle_namcao(sample_size):
    proportions = []
    for i in range(1000):
        sample = np.random.choice(heights, sample_size)
        num_taller = np.sum(sample > 180)
        proportion = num_taller / sample_size
        proportions.append(proportion)
    return proportions

# Mô phỏng lấy mẫu có kích thước 100 và vẽ histogram tỷ lệ người cao của các mẫu
proportions_100 = get_tyle_namcao(100)
plt.hist(proportions_100, bins=20,width=0.01)
plt.title("Kích thước mẫu: 100")
plt.xlabel("Tỷ lệ người cao")
plt.ylabel("Số lượng mẫu")
plt.show()

# Mô phỏng lấy mẫu có kích thước 500 và vẽ histogram tỷ lệ người cao của các mẫu
proportions_500 = get_tyle_namcao(500)
plt.hist(proportions_500, bins=20,width=0.005)
plt.title("Kích thước mẫu: 500")
plt.xlabel("Tỷ lệ người cao")
plt.ylabel("Số lượng mẫu")
plt.show()
```

### Bài 6: Khảo sát phương sai mẫu

Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram đại lượng phương sai của các mẫu (Phương sai bằng bình phương độ lệch chuẩn)

```python

heightss = np.random.randint(120,201,size=10000)
# Tính đại lượng phương sai của các mẫu lấy từ dân số
def get_sample_variances(sample_size):
    variances = []
    for i in range(1000):
        sample = np.random.choice(heightss, sample_size,replace=True)
        variance = np.var(sample)
        variances.append(variance)
    return variances

# Mô phỏng lấy mẫu có kích thước 100 và vẽ histogram đại lượng phương sai của các mẫu
variances_100 = get_sample_variances(100)
plt.hist(variances_100, bins=20, width=11)
plt.title("Kích thước mẫu: 100")
plt.xlabel("Phương sai")
plt.ylabel("Số lượng mẫu")
plt.show()
```

### Bài 7:  Bài Tập Ước Lượng 

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
from scipy.stats import norm

data = np.array([2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8])
n = len(data)
alpha = 0.05

sigma = 0.3  # độ lệch chuẩn của quần thể
se = sigma / np.sqrt(n)  # độ lệch chuẩn của phân phối mẫu trung bình
z_alpha_2 = norm.ppf(1 - alpha / 2)  # giá trị z_{alpha/2} tương ứng

lower_bound = data.mean() - z_alpha_2 * se
upper_bound = data.mean() + z_alpha_2 * se

print(f"Khoảng tin cậy [{lower_bound:.4f}, {upper_bound:.4f}]")
```

```python
from scipy.stats import t
t_alpha_2 = t.ppf(1 - alpha / 2, df=n-1)  # giá trị t_{alpha/2, n-1} tương ứng
ses = data.std(ddof=1) / np.sqrt(n)  # độ lệch chuẩn mẫu của phân phối trung bình

lower_bound = data.mean() - t_alpha_2 * ses
upper_bound = data.mean() + t_alpha_2 * ses

print(f"Khoảng tin cậy [{lower_bound:.4f}, {upper_bound:.4f}]")
```
