Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

````python
# Your code here
# Phát sinh dữ liệu x từ 0 đến pi
x = np.linspace(0, np.pi, 100)

# Tính giá trị của hàm sin(x) và cos(x)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Vẽ đồ thị hàm sin(x) màu xanh dương
plt.plot(x, y_sin, label='sin(x)', color='blue')

# Vẽ đồ thị hàm cos(x) màu cam, nét đứt
plt.plot(x, y_cos, label='cos(x)', color='orange', linestyle='--')

# Đặt các nhãn và tiêu đề
plt.xlabel('x')
plt.ylabel('y')
plt.title('Biểu đồ của hàm sin(x) và cos(x)')
plt.legend()  # Hiển thị chú thích
plt.grid(True)  # Hiển thị lưới
plt.show()
````

````python
# Your code here
import numpy as np
import matplotlib.pyplot as plt

# Phát sinh mẫu 10000 số ngẫu nhiên theo phân phối chuẩn
data = np.random.normal(size=10000)

# Tạo một figure và một set of subplots
fig, axs = plt.subplots(2)

# Vẽ histogram của mẫu số ngẫu nhiên
axs[0].hist(data, bins=50, color='purple')
axs[0].set_title('Histogram')

# Vẽ boxplot của mẫu số ngẫu nhiên với tham số vert=False
axs[1].boxplot(data, vert=False)
axs[1].set_title('Boxplot')

# Hiển thị đồ thị
plt.tight_layout()
plt.show()
````

## SEABORN



Seaborn là một thư viện hỗ trợ trực quan hóa được viết dựa trên thư viện `matplotlib`. Trong bài tập này, bạn hãy tự tìm hiểu thư viện `seaborn` và trực quan hóa dữ liệu tiền tip tại một nhà hàng.

Bạn có thể dùng đoạn mã sau để đọc dữ liệu có sẵn trong `seaborn`

```python
import seaborn as sns

df = sns.load_dataset("tips")
df.head()
```
Hoặc bạn có thể đọc dữ liệu chứa trong file `tips.csv` chứa thông tin tiền tip của khách hàng tại một nhà hàng. Dữ liệu có các thông tin sau:

- `total_bill`: tổng hoá đơn (USD)
- `tip`: tiền tip/ tiền boa (USD)
- `sex`: giới tính người thanh toán hóa đơn
- `smoker`: có hút thuốc hay không
- `day`: ngày trong tuần
- `time`: thời gian dùng bữa(buổi sáng, buổi trưa, buổi tối)
- `size`: số lượng người tham gia

Bạn hãy lựa chọn loại đồ thị phù hợp để trả lời những câu hỏi sau:

### 1. Vẽ đồ thị thể hiện tổng tiền tip theo ngày

````python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

tip_by_day = df.groupby('day')['tip'].sum().reset_index()
sns.barplot(x='day', y='tip', data=tip_by_day, palette='muted')
plt.title('Tổng Tiền Tip Theo Ngày')
plt.xlabel('Ngày')
plt.ylabel('Tổng Tiền Tip (USD)')
plt.show()

````

### 2. Vẽ đồ thị thể hiện tổng tiền tip theo thời gian dùng bữa (time)


````python
tip_by_time = df.groupby('time')['tip'].sum().reset_index()
sns.barplot(x='time', y='tip', data=tip_by_time, palette='muted')
plt.title('Tổng Tiền Tip Theo Thời Gian Dùng Bữa')
plt.xlabel('Thời Gian Dùng Bữa')
plt.ylabel('Tổng Tiền Tip (USD)')
plt.show()
````

### 3. Vẽ đồ thị histogram của tổng hoá đơn


````python
sns.histplot(df['total_bill'], bins=20, kde=True, color='skyblue')
plt.title('Phân phối của Tổng Hoá Đơn')
plt.xlabel('Tổng Hoá Đơn (USD)')
plt.ylabel('Số Lượng')
plt.show()
````

### 4. Vẽ đồ thị boxplot của tổng hoá đơn


````python
sns.boxplot(x='total_bill', data=df, color='skyblue')
plt.title('Boxplot của Tổng Hoá Đơn')
plt.xlabel('Tổng Hoá Đơn (USD)')
plt.show()
````

### 5. Vẽ đồ thị boxplot so sánh tổng hóa đơn giữa các ngày trong tuần


````python
sns.boxplot(x='day', y='total_bill', data=df, palette='muted')
plt.title('Boxplot so sánh Tổng Hoá Đơn giữa các ngày trong tuần')
plt.xlabel('Ngày trong tuần')
plt.ylabel('Tổng Hoá Đơn (USD)')
plt.show()
````

### 6. Vẽ đồ thị để kiểm tra sự liên quan giữa tổng hoá đơn và tiền tip. (`scatter plot`)


````python
sns.scatterplot(x='total_bill', y='tip', data=df, color='skyblue')
plt.title('Sự Liên Quan giữa Tổng Hoá Đơn và Tiền Tip')
plt.xlabel('Tổng Hoá Đơn (USD)')
plt.ylabel('Tiền Tip (USD)')
plt.show()
````

### 7. Tìm hiểu và sử dụng hàm `jointplot` trong seaborn để khảo sát quan hệ giữa hai thuộc tính `total_bill` và `tip`. Theo bạn hai thuộc tính này có tương quan với nhau hay không?


````python
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
plt.show()
````

### 8. Tìm hiểu và sử dụng hàm `pairplot` trong seaborn để trực quan bộ dữ liệu. Theo bạn, trong bộ dữ liệu có những thuộc tính nào liên quan với nhau không?



````python
sns.pairplot(df)

plt.show()
````

### 9. Vẽ hai đồ thị histogram cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`



````python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Vẽ histogram cho thời gian dùng bữa Lunch
sns.histplot(df[df['time'] == 'Lunch']['total_bill'], ax=axs[0], kde=True, color='skyblue')
axs[0].set_title('Tổng Hoá Đơn - Lunch')
axs[0].set_xlabel('Tổng Hoá Đơn (USD)')
axs[0].set_ylabel('Số Lượng')

# Vẽ histogram cho thời gian dùng bữa Dinner
sns.histplot(df[df['time'] == 'Dinner']['total_bill'], ax=axs[1], kde=True, color='salmon')
axs[1].set_title('Tổng Hoá Đơn - Dinner')
axs[1].set_xlabel('Tổng Hoá Đơn (USD)')
axs[1].set_ylabel('Số Lượng')

# Hiển thị đồ thị
plt.tight_layout()
plt.show()
````

### 10. Vẽ hai đồ thị boxplot cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`

````python
# Tạo hai subplot
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Vẽ boxplot cho thời gian dùng bữa Lunch
sns.boxplot(x='time', y='total_bill', data=df[df['time'] == 'Lunch'], ax=axs[0], palette='pastel')
axs[0].set_title('Tổng Hoá Đơn - Lunch')
axs[0].set_xlabel('Thời Gian Dùng Bữa')
axs[0].set_ylabel('Tổng Hoá Đơn (USD)')

# Vẽ boxplot cho thời gian dùng bữa Dinner
sns.boxplot(x='time', y='total_bill', data=df[df['time'] == 'Dinner'], ax=axs[1], palette='muted')
axs[1].set_title('Tổng Hoá Đơn - Dinner')
axs[1].set_xlabel('Thời Gian Dùng Bữa')
axs[1].set_ylabel('Tổng Hoá Đơn (USD)')

# Hiển thị đồ thị
plt.tight_layout()
plt.show()
````
