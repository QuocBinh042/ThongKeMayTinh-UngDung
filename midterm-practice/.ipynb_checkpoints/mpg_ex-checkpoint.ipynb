### Câu 4: (2 điểm)

a, (1.5 điểm) Vẽ biểu đồ tăng trưởng về dân số của 10 quốc gia Đông Nam Á. 

b, (0.5 điểm) Dựa vào hình vẽ bạn hãy cho biết quốc gia nào có tốc độ tăng trưởng nhanh nhất?

---

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Thông số của phân phối chuẩn
mu = 172  # Trung bình
sigma = 5  # Độ lệch chuẩn

# a) Xác suất chọn ngẫu nhiên một bạn gọi là lùn (nhỏ hơn 162(cm))
prob_lun = norm.cdf(162, mu, sigma)
print("a) Xác suất chọn ngẫu nhiên một bạn gọi là lùn (nhỏ hơn 162cm):", prob_lun)

# b) Xác suất chọn ngẫu nhiên một thanh niên nam cao lớn (> 182cm)
prob_cao_lon = 1 - norm.cdf(182, mu, sigma)
print("b) Xác suất chọn ngẫu nhiên một thanh niên nam cao lớn (> 182cm):", prob_cao_lon)

# c) Xác suất chọn người tầm thước (trong khoảng 165-182cm)
prob_tam_thuoc = norm.cdf(182, mu, sigma) - norm.cdf(165, mu, sigma)
print("c) Xác suất chọn người tầm thước (trong khoảng 165-182cm):", prob_tam_thuoc)

# d) Chuẩn thấp bé của thanh niên nam ở Hàn Quốc (cm)
thap_be = norm.ppf(0.06, mu, sigma)
print("d) Chuẩn thấp bé của thanh niên nam ở Hàn Quốc (cm):", thap_be)

# e) Cao lớn có ngưỡng chiều cao là bao nhiêu (cm)?
cao_lon = norm.ppf(0.91, mu, sigma)
print("e) Cao lớn có ngưỡng chiều cao là bao nhiêu (cm):", cao_lon)

# f) Vẽ biểu đồ phân bố mô tả phân bố chuẩn
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, norm.pdf(x, mu, sigma), label='Phân bố chuẩn')
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Xác suất')
plt.title('Phân bố chiều cao của nam thanh niên Hàn Quốc')
plt.legend()
plt.grid(True)
plt.show()
```
