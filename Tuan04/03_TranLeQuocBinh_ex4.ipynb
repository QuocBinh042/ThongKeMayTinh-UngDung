**Bài 04:** Mô phỏng tung một con xúc sắc cân đối đồng chất 5000 lần. Dựa vào giá trị mô phỏng, bạn hãy tìm các giá trị xác suất dưới đây:

a. Xác suất để  số chấm xuất hiện là 4

b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4

c. Giả sử biết số chấm xuất hiện lớn hơn hoặc bằng 4. Hãy tìm xác suất để mặt 6 chấm xuất hiện.

```python
import numpy as np

# Số lần tung xúc sắc
n_tosses = 5000

# Mô phỏng tung xúc sắc và đếm số lần mặt xuất hiện
dice_results = np.random.randint(1, 7, n_tosses)

# a. Xác suất để số chấm xuất hiện là 4
prob_4 = np.sum(dice_results == 4) / n_tosses
print("a. Xác suất để số chấm xuất hiện là 4:", prob_4)

# b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4
prob_greater_than_equal_4 = np.sum(dice_results >= 4) / n_tosses
print("b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4:", prob_greater_than_equal_4)

# c. Giả sử biết số chấm xuất hiện lớn hơn hoặc bằng 4. 
# Tìm xác suất để mặt 6 chấm xuất hiện.
# Tính số lần xuất hiện của mặt 6 trong các lần số chấm >= 4
num_6_after_4 = np.sum(dice_results[dice_results >= 4] == 6)
# Xác suất được tính bằng số lần xuất hiện mặt 6 chia cho tổng số lần số chấm >= 4
prob_6_given_greater_than_equal_4 = num_6_after_4 / np.sum(dice_results >= 4)
print("c. Xác suất để mặt 6 chấm xuất hiện khi số chấm lớn hơn hoặc bằng 4:", prob_6_given_greater_than_equal_4)
```

**Bài 05:** Tạo 5000 số ngẫu nhiên có phân phối nhị thức với n=50, p=0.7 Bạn Hãy

a. Tìm các giá trị thống kê: min, max, Q1, Q2, Q3

b. Vẽ đồ thị boxplot

c. Vẽ các đồ thị: tần số, tần suất

```python
# a. Tạo dữ liệu ngẫu nhiên theo phân phối nhị thức
n = 50
p = 0.7
random_data = np.random.binomial(n, p, 5000)

# b. Tính các giá trị thống kê: min, max, Q1, Q2 (median), Q3
min_val = np.min(random_data)
max_val = np.max(random_data)
Q1 = np.percentile(random_data, 25)
Q2 = np.percentile(random_data, 50)  # median
Q3 = np.percentile(random_data, 75)

print("Min:", min_val)
print("Max:", max_val)
print("Q1:", Q1)
print("Q2 (Median):", Q2)
print("Q3:", Q3)

# b. Vẽ đồ thị boxplot
plt.boxplot(random_data)
plt.title("Boxplot")
plt.show()

# c. Vẽ biểu đồ tần số
plt.hist(random_data, bins=20, density=False, alpha=0.75)
plt.title("Histogram - Frequency")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

# c. Vẽ biểu đồ tần suất
plt.hist(random_data, bins=20, density=True, alpha=0.75)
plt.title("Histogram - Relative Frequency")
plt.xlabel("Values")
plt.ylabel("Relative Frequency")
plt.show()
```
